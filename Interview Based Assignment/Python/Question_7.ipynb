```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['mass']=df['mass'].fillna(df['mass'].median()) # mass column have 28 null value so here replaced with median value
```

```python
df['year']=df['year'].fillna(df['year'].mode()[0]) # year column have only one null value so here replaced with most_frequency value
```

```python
df.info()
```

### Get all the Earth meteorites that fell before the year 2000

```python
split_year=lambda x:int(x.split('-')[0])
split_year('1880-01-01')
```

```python
first_q=df[df['year'].apply(split_year)<2000]['Name of Earth Meteorite'].to_list()
print('Get all the Earth meteorites that fell before the year 2000:\n',first_q,len(first_q))
```

### Get all the earth meteorites co-ordinates who fell before the year 1970

```python
split_year=lambda x:int(x.split('-')[0])
split_year('1880-01-01')
```

```python
pd.set_option('display.max_rows', None)
second_q=df[df['year'].apply(split_year)<1970][['Name of Earth Meteorite','point coordinates']]
second_q.shape
```

### Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
third_q=df[df['mass']>10000]['Name of Earth Meteorite'].to_list()
print('Get all those whose mass was more than 10000kg:\n',third_q,len(third_q))
```
