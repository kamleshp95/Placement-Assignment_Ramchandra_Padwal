### Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.

#### Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from collections import defaultdict
```

```python
# Function to count the occurrences of verbs, nouns, pronouns, and adjectives
def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Perform part-of-speech tagging on the words
    tagged_words = pos_tag(words)

    # Count the occurrences of verbs, nouns, pronouns, and adjectives
    counts = defaultdict(int)
    for word, tag in tagged_words:
        if tag.startswith('V'):
            counts['verbs'] += 1
        elif tag.startswith('N'):
            counts['nouns'] += 1
        elif tag == 'PRP' or tag == 'PRP$':
            counts['pronouns'] += 1
        elif tag.startswith('JJ'):
            counts['adjectives'] += 1

    return counts
```

```python
# Main program
def main():
    # Test case 1
    phrase = "The quick brown fox jumps over the lazy dog."
    counts = count_pos_tags(phrase)
    print("Test case 1: ",phrase)
    print("Counts for Test Case 1:")
    print(counts)

    # Test case 2
    paragraph = "I saw a beautiful sunset on the beach. The waves were crashing against the shore while seagulls flew overhead."
    counts = count_pos_tags(paragraph)
    print("Test case 2: ",paragraph)
    print("Counts for Test Case 2:")
    print(counts)

    # test case 3
    sentence = "She loves to read books and write stories."
    counts = count_pos_tags(sentence)
    print("Test case 3: ",sentence)
    print("Counts for Test Case 3:")
    print(counts)


# Run the program
if __name__ == "__main__":
    # Download the necessary resources for NLTK
    # nltk.download('punkt')
    # nltk.download('averaged_perceptron_tagger')

    # Call the main function
    main()
```
